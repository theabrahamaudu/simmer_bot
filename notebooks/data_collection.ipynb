```python
import os
import datetime
import MetaTrader5 as mt5
import pandas as pd
```

```python
login = os.environ.get("LOGIN")
password = os.environ.get("PASSWORD")
server = os.environ.get("SERVER")
 
# runs MetaTrader5 client
# MetaTrader5 platform must be installed
mt5.initialize()
 
# log in to trading account
mt5.login(login, password, server) 
```

```python
symbol = 'EURUSD'
timeframe = 1 # integer value representing minutes
start_date = datetime.datetime(year=2023, month=1, day=1).strftime("%Y-%m-%d %H:%M:%S.%f")
start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d %H:%M:%S.%f")
end_date = datetime.datetime.now()
 
# bars = mt5.copy_rates_from_pos(symbol, timeframe, start_bar, num_bars)
bars = mt5.copy_rates_range(symbol, mt5.TIMEFRAME_M15, start_date, end_date)
```

```python
print(type(start_date), type(end_date))
print(type(bars))
```

```python
# create DataFrame out of the obtained data
rates_frame = pd.DataFrame(bars)
# convert time in seconds into the datetime format
rates_frame['time']=pd.to_datetime(rates_frame['time'], unit='s')
```
